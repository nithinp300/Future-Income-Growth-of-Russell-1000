```python
cwd = 'drive/My Drive/data mining project/'
unlabeled_cwd = 'drive/My Drive/data mining project/unlabeled data/'
result_cwd = cwd + 'final result/'
sector_list = pd.read_csv('drive/My Drive/data mining project/sector_list.csv', header=None)

regressions = ['radius_nearest_neignbor_regression', 'sv_regression', 'rlp_regression', 'RANSAC_regression', 'lasso_regression', 'linear_regression']
```

# Read in result files

## read mean square errors

```python
init_mse = pd.read_csv(cwd + 'Initial_MSE.csv', header=None, names=['industry', 'mse'])

rnr_mse = pd.read_csv(cwd + regressions[0] + '_result/mse.csv', header=None)
svr_mse = pd.read_csv(cwd + regressions[1] + '_result/mse.csv', header=None)
rlp_mse = pd.read_csv(cwd + regressions[2] + '_result/mse.csv', header=None)
ran_mse = pd.read_csv(cwd + regressions[3] + '_result/mse.csv', header=None)
lasso_mse = pd.read_csv(cwd + regressions[4] + '_result/mse.csv', header=None)
linear_mse = pd.read_csv(cwd + regressions[5] + '_result/mse.csv', header=None)

mse = [rnr_mse, svr_mse, rlp_mse, ran_mse, lasso_mse, linear_mse]
init_mse.head(30)
```

## determine the best regressor for each industry

```python
report_list = []
```

```python
for industry in sector_list.iterrows():
  ind_name = industry[1][0]
  ind_index = industry[0]
  min_mse = init_mse['mse'][ind_index]
  mse_initial = min_mse
  best_reg = None
  
  # loop through all the regressors to find the one with minimum mse
  reg_idx = 0
  for reg in regressions:
    reg_mse = mse[reg_idx][0][ind_index]
    if min_mse > reg_mse:
      min_mse = reg_mse
      best_reg = regressions[reg_idx]
    reg_idx += 1

  result = pd.read_csv(cwd + best_reg + '_result/' + ind_name + '.csv', header=None)
  unlabel_data = pd.read_csv(unlabeled_cwd + ind_name + '.csv')
  unlabel_data['future_net_income_growth'] = result[0]
  unlabel_data = unlabel_data[['symbol', 'industry', 'period_end_date', 'future_net_income_growth']].sort_values(by='future_net_income_growth', ascending=False)
  
  unlabel_data.to_csv(result_cwd + ind_name + '_result.csv', index=None)

  # add a report
  report_list.append([ind_name, best_reg, min_mse, mse_initial])
```

## add a report of regressors

```python
report_df = pd.DataFrame(report_list, columns=['Industry', 'Best Regressor', 'MSE with Best Reg', 'MSE with naive predict'])
report_df.to_csv(cwd + 'Error_Report.csv', index=None)
report_df.head(30)
```
